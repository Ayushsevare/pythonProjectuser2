```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Ai survey.csv")
```

```python
data.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = ['How important do you consider the explainability of AI systems in gaining your trust as a user?',
       ' how confident are you in your ability to understand the decisions made by\nAI systems without transparent explanations?',
       'Have you ever felt uncomfortable or hesitant to use an AI-driven product or service due to a lack\nof transparency in its decision-making process?',
       'Do you believe that AI developers and providers have a responsibility to ensure that their\nsystems are explainable and transparent to users?',
       'What types of explanations or information would you find most helpful in understanding the\ndecisions made by AI systems (e.g., feature importance, decision rationale, model confidence)?',
       'Have you ever experienced a situation where an AI system provided an explanation for its\ndecision that you found unsatisfactory or insufficient?',
       'How would you rate your level of trust in AI systems that offer transparent explanations\ncompared to those that do not?',
       'Do you think regulatory bodies should mandate minimum standards for explainability and\ntransparency in AI systems, similar to data privacy regulations like GDPR?',
       'What concerns, if any, do you have about the potential negative consequences of over-reliance\non AI systems that lack transparency or explainability?',
       'In your opinion, what measures can AI developers and providers take to improve the\nexplainability and trustworthiness of their systems in the eyes of users and stakeholders?']
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_cols = encoder.fit_transform(data[categorical_cols])
feature_names = encoder.get_feature_names_out(categorical_cols)
encoded_cols = pd.DataFrame(encoded_cols, columns=feature_names)
data.drop(categorical_cols, axis=1, inplace=True)
data = pd.concat([data, encoded_cols], axis=1)
```

```python
stop_words = set(stopwords.words('english'))
words = word_tokenize(qualitative_text.lower())
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
```

```python
word_freq = nltk.FreqDist(filtered_words)
print("Top 10 Most Common Words:")
print(word_freq.most_common(10))
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(qualitative_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
