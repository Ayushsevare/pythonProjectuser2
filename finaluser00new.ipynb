```python
import plotly.express as px
```

```python
path= pd.read_csv('/content/final_map_user_new.csv')
```

```python
fig = px.choropleth(
    path,
    locations='state',
    locationmode='USA-states',
    color='registeredUsers',
    scope='usa'
)
```

```python
fig.show()
```

```python
features = ['registeredUsers', 'appOpens']
```

```python
kmeans = KMeans(n_clusters=3)
path['cluster'] = kmeans.fit_predict(path[features])
```

```python

fig = px.scatter(path, x='registeredUsers', y='appOpens', color='cluster', hover_data=['state'])
fig.show()
```
